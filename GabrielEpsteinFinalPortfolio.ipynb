# **Let's Jump In!**


**Importing Libraries:** Below, I am importing some of the libraries I will be using to analyze this data. So far, I am using Pandas and NumPy libraries.

```python
#Importing pandas and numpy for use in analysis
import pandas as pd 
import numpy as np

pd.set_option("display.max_rows", None, "display.max_columns", None)
```

 **Importing Data:** Here, I am importing the dataset that I found from [Data.Gov](https://data.gov), the "home of the U.S. Government's Open Data." Then I will use the .head() function to visually verify that the dataset seems to have been imported without problems


```python
#Reading the csv and displaying the first n (5 by default) rows with .head()
whole_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Science Final Portfolio Files/Accidental_Drug_Related_Deaths_2012-2021.csv')
whole_df.head(10)
```

Next, we will take a look at what types of objects are contained within the dataset

```python
#Getting the dtypes
whole_df.dtypes
```

Next, we will do some clean up. I want to look at my data over time, so I will first convert the dates of type object into datetimes, sort said datetimes, and then set my index as the 'Date' to have my dataframe indexed by date and chronologically ordered.

```python
#DataFrame in its originally indexed state
whole_df.head(5)
```

Now, I will convert the 'Date' datatype into a datetime. Once I've done so, I will sort said datetime values, then index the dataframe by date.

```python
whole_df['Date'] = pd.to_datetime(whole_df['Date'])
#whole_df.info() used .info() in order to ensure that Date has been changed to dtype datetime instead of object

whole_df.sort_values(by='Date', inplace = True) #Sorting the 'Date' entries, and making sure the change is not in a temporary new dataframe with inplace = True
#whole_df.head(5)
df_by_date = whole_df.set_index('Date') #creating a new dataframe with the sorted 'Date' entries as the index
df_by_date.head(5)
```

**Displaying Data:** As an introduction to graphing and displaying the data visually, I will show the histogram of deaths by age and the barplot of deaths by sex.

```python
#Getting some basic numerical information about the entries with 'Age' attributes
df_by_date['Age'].describe()

#Plotting the histogram, setting the number of bins, and enabling the density as the y axis
df_by_date.Age.plot.hist(bins = 25, density = True, xlabel = 'Age', sharex = True)
df_by_date.Age.plot.density(xlim = (5, 90), xlabel = 'Age', sharex = True)
```

```python
#Since Sex is one of two values, I will get a count of each option ('Male', 'Female') to make a barlpot
sex_counts = df_by_date.Sex.value_counts()
sex_counts.plot.bar(ylabel = 'Count', xlabel = 'Sex', color = ['blue', 'pink'])
```
